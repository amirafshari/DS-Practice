```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import pandas_ta as ta

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('S_Mobarakeh.Steel.csv')
```

# 1.PP

```python
df = data.rename(columns={'<DTYYYYMMDD>': 'date', '<CLOSE>': 'close', '<FIRST>': 'open', \
                          '<LOW>': 'low', '<HIGH>': 'high', '<VOL>': 'volume', '<VALUE>': 'market_value', \
                          '<OPEN>': 'adj_close', '<OPENINT>': 'open_interest' })


index = pd.to_datetime(df.date, format='%Y%m%d')

df.drop(['<LAST>', '<TICKER>', '<PER>', 'date', 'market_value', 'open_interest', 'volume'], axis=1, inplace=True)

df.index = index
```

```python
'''
df[::-1].ta.rsi(append=True)[::-1]
df[::-1].ta.ema(append=True)[::-1]
df.dropna(how='any', inplace=True)
'''
```

```python
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df), index=df.index, columns=df.columns)
```

# 2.PCA

```python
pca = PCA(2)
dataframe = pd.DataFrame(pca.fit_transform(scaled_df), columns=['PC1', 'PC2'], index=df.index)
pca.explained_variance_ratio_
```

# 3.Viz

```python
fig = px.line(scaled_df, x=dataframe.index, y='close')
fig
```

```python
fig = px.line(dataframe, x=dataframe.index, y='PC1')
fig
```
